```python
# Import pandas
import pandas as pd
```

```python
# Create a DataFrame with balance.txt
df = pd.read_csv('balance.txt', delim_whitespace=True)
```

```python
#Compare the average income based on ethnicity
average_income_by_ethnicity = df.groupby('Ethnicity')['Income'].mean()
print(average_income_by_ethnicity)
```

```python
# On average, do married or single people have a higher balance
average_balance_by_marital_status = df.groupby('Married')['Balance'].mean()
print(average_balance_by_marital_status)
print("On average, the single people have a higher balance")
```

```python
# what is the highest income in our dataset
highest_income = df['Income'].max()
print("The highest income in the dataset is:", highest_income)
```

```python
# what is the lowest income in our dataset
lowest_income = df['Income'].min()
print("The lowest income in the dataset is:", lowest_income)
```

```python
#How many cards do we have recorded in our dataset
Cards_num = df.Cards.sum()
print("The number of cards recorded in the dataset is: ", Cards_num)
```

```python
# How many females VS how many males
Female_count=df[df['Gender']=='Female'].count()
Male_count=df[df['Gender']=='Male'].count()
print("Number of females:", Female_count)
print("Number of females:", Male_count)
```
