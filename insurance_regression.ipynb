```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Import the data
data = pd.read_csv("insurance.csv")
data.head()
```

```python
# The iloc function allows use to select rows from the dataframe for plotting.  
x = data.iloc[:,0].values
print(x)
y = data.iloc[:,6].values
# To alloow compatability, the inputs must be reshaped
x=x.reshape(-1, 1)
y=y.reshape(-1, 1)
```

```python
insurance_model = LinearRegression()

# to create the model, we use fit(x,y)
insurance_model.fit(x,y)

y_pred = insurance_model.predict(x)
plt.scatter(x,y,color = 'b')
plt.plot(x,insurance_model.predict(x),color = 'r')
plt.title('Age v/s Charges')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```
